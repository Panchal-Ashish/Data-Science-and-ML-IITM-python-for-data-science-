```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv("Toyota_corolla_modified.csv", na_values = ['??','????'])
print(cars_data.info())
```

## Removing missing values from dataframe
#### Dataframe.dropna(axis = (0 or 1), how = ('any' or 'all'), inplace = (True or false))
to remove missing values for further operations

#### axis (default: 0)
0: Drop rows (or index) having missing values

1: Drop columns having missing values
#### how (default: 'any')
'any': if any NA values present, drop that entire row or column

'all': if all values NA, drop that row or column
#### inplace (default: False)
True: do the operation and store inplace(same variable)

False: if we have created a new variable or copy, then store over in that copy , not original

```python
cars_data.dropna(axis =0, how ="any", inplace =True)
print(cars_data.info())
```

# SCATTER PLOT

#### sns.set(style = ' ', rc={'figure.figsize':(x value(eg. 11.7), y value(eg. 8.27))})
available options: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'


#### sns.regplot( x = X axis dataseries, y = Y axis dataseries, fit_reg = True/False (default True), marker = " ")
fit_reg : Estimate and plot a regression line from the data

marker : If you want to change the marker, specify it here

```python
sns.set(style = 'whitegrid')
sns.regplot(x = cars_data["Age"], y = cars_data['Price'], fit_reg = False)
```

```python
sns.set(style = 'darkgrid')
sns.regplot(x = cars_data["Age"], y = cars_data['Price'], fit_reg = True, marker = "*").set_title("Scatter plot of Age vs Price")
```

### Scatter plot of 'A' vs 'B' by 3rd variable 'C'
using 'hue' parameter, we cam add the 3rd variavle

#### sns.lmplot(x= , y= , data= Dataframe, fit_reg= , hue= 3rd variable (column label), legend= True/False, palette = )
 
x : dataseries on X axis... only the column label

y : dataseries on Y axis... only the column label

fit_reg.... same as above

data : Dataframe from where the x, y, and 3rd variable are specified

hue : 3rd variable (column label)

palette : can specify the colors inside a list or a tuple.... or else, use the inbuilt color palettes using 'Set1/2/3/4'

```python
sns.set(style = 'darkgrid')
sns.lmplot(x = "Age", 
            y = 'Price',
            data = cars_data,
            hue = "FuelType", 
            legend = True,
           palette = ["red","green","orange",],
            fit_reg = True,)
```

```python
sns.set(style = 'darkgrid')
sns.lmplot(x = "Age", 
            y = 'Price',
            data = cars_data,
            hue = "Doors", 
            legend = True,
           palette = ["red","green","orange","blue"],
            fit_reg = True,)
```

```python
#cars_data["Doors"] = cars_data["Doors"].map({'three':3,3:3,'four':4,4:4,'five':5,5:5})
cars_data["Doors"].replace({'three':3,3:3,'four':4,4:4,'five':5,5:5}, inplace = True)
sns.lmplot(x = "Age", 
            y = 'Price',
            data = cars_data,
            hue = "Doors", 
            legend = True,
           palette = ["red","green","orange","blue"],
            fit_reg = False)
```

## BOX AND WHISKERS PLOT

### sns.boxplot(x (or y) = Dataseries)
If you specify x axis, then the plot will be along X axis

if you specify y axis, then the plot will be along Y axis

```python
sns.boxplot(y = cars_data["Price"])
```

Top most line: MAX (excluding outliers)... Top whisker          [Max price is around 17500]

Bottom most line: MIN (excluding outliers)...Bottom whisker      [Min price is around 5000]

Q1. Bottom box line (Quantile 1): 25% data (here price) lies below this line      [25% data lies below 8000]

Q2. Middle Box line (Quantile 2): Median... 50% data lies below this line         [50% data lies below 10000]... Median

Q3. Top box line (Quantile 3): 75% data lies below this line                      [75% data lies below 12500]

all data outside the 
top and bottom whiskers are called 'Outliers'

MAX =  Q3 + (1.5 * IQR)

MIN = Q1 - (1.5 * IQR)

Where IQR = interquartile range

```python
cars_data["Price"].describe()
```

# Multiple plots in single window

## Box - whiskers plot and histogram 
step 1: split the window
#### fig,(axis_box, axis_hist) = plt.subplots(no. of rows (int), gridspec_kw = {"height_ratios": (height ratio of 1st plot, height ratio of 2nd plot)})
#### matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)

sharex / sharey : whether the scale of that axis is shared or not

NOTE: it is Matlotlib... not seaborn



step 2: Create the plots
#### sns.plotname(Dataseries, ax = axis of that plot)

```python
fig,(axis_box,axis_hist) = plt.subplots(2,gridspec_kw = {"height_ratios": (0.25,0.75)},)
sns.boxplot(cars_data["Price"], ax= axis_box)
sns.histplot(cars_data["Price"], ax = axis_hist, kde = False)
```

```python
fig,(axis_box,axis_hist) = plt.subplots(nrows = 1, 
                                        ncols =2,
                                        gridspec_kw = {"width_ratios": (0.25,0.75)},
                                       sharex = True,
                                       sharey = False)
sns.boxplot(cars_data["Price"], ax= axis_box)
sns.histplot(cars_data["Price"], ax = axis_hist, kde = False)
```

# PAIR WISE PLOTS
CREATES ALL POSSIBLE PLOTS FOR THE GIVEN VARIABLE

used to plot pairwise relationships in a dataset

creates SCATTERPLOTS for joint relationships (different variables on X and Y axis)

creates HISTOGRAM for univariate distribution (Same variable on X and Y axis)

### sns.pairplots(Dataframe, kind = 'scatter' (type of plot), hue= 'variable'(column label))

kind : "scatter", "kde", "reg", "hist" 

```python
sns.pairplot(cars_data, kind = 'scatter')
plt.show()
```

```python
sns.pairplot(cars_data, kind = 'scatter', hue = 'Automatic')
plt.show()
```

```python
sns.pairplot(data = cars_data, kind = 'scatter', hue = 'FuelType', palette = ["red","blue","green"])
sns.set(style = 'darkgrid')
plt.show()
```

```python
sns.pairplot(cars_data, kind = 'hist', hue = 'Automatic')
plt.show()
```

```python
sns.pairplot(cars_data, kind = 'reg', hue = 'Automatic')
plt.show()
```

```python
sns.pairplot(cars_data, kind = 'scatter', hue = 'Doors', palette = ["red","yellow","blue","green"])
```
